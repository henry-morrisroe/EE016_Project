```python
import pandas as pd
import numpy as np # Added numpy import

# Load data
df = pd.read_csv("/content/Titanic-Dataset.csv")

# Check for missing values
print(df.isnull().sum())

# Fill missing values with median or average.
#If the number of null rows is not large, you can drop those rows.
df["Age"].fillna(df["Age"].median(),inplace=True)
df["Cabin"].fillna(df["Cabin"].mode(),inplace=True)
df["Embarked"].fillna(df["Embarked"].mode(),inplace=True)

from sklearn.preprocessing import LabelEncoder, StandardScaler
# Encode categorical columns
for col in df.select_dtypes(include="object"): # Use 'object' for categorical strings
  df[col] = LabelEncoder().fit_transform(df[col])

for col in df.select_dtypes(include="float64"): # Use 'object' for categorical strings
  df[col] = LabelEncoder().fit_transform(df[col])

# Feature scaling
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])


# Check for missing values
print()
print(df.isnull().sum())
```
